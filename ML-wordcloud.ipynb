```python
#This code reads the data into a pandas data frame:

import pandas as pd
import gzip
import json

def parse(path):
  g = gzip.open(path, 'rb')
  for l in g:
    yield json.loads(l)

def getDF(path):
  i = 0
  df = {}
  for d in parse(path):
    df[i] = d
    i += 1
  return pd.DataFrame.from_dict(df, orient='index')

df = getDF('/content/reviews_Amazon_Instant_Video_5.json.gz')
df1 = getDF('/content/reviews_Digital_Music_5.json.gz')
df_office = getDF('reviews_Office_Products_5.json.gz')
df_beauty = getDF('reviews_Beauty_5.json.gz')

```

# **Amazon Instant Video Review** 

```python
text = " ".join(review for review in df.reviewText)
print ("There are {} words in the combination of all review.".format(len(text)))
```

```python
stopwords = set(STOPWORDS)
# Generate a word cloud image
wordcloud = WordCloud(stopwords=stopwords, background_color="white",max_font_size=100, max_words=10).generate(text)

# Display the generated image:
# the matplotlib way:
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.figure(figsize=[20,10])
plt.show()
```
